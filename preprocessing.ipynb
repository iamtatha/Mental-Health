```python
df = pd.read_csv("https://raw.githubusercontent.com/iamtatha/Data_Science/master/survey.csv")
headers = list(df.keys())
```

```python
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "Cis Male", "cis male"]
trans_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]           
female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]


for i in range(len(df['Gender'])):
  text = df['Gender'][i].lower()
  if text in male_str:
    change = 'M'
  elif text in female_str:
    change = 'F'
  else:
    change  = 'T'
  df['Gender'][i] = change
```

```python
one_hot_cols = ['work_interfere', 'benefits', 'care_options', 'wellness_program', 'seek_help', 'anonymity', 'leave', 'mental_health_consequence',
 'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview','mental_vs_physical',]

label_cols = ['Age', 'Gender', 'Country', 'state', 'self_employed', 'family_history', 'no_employees', 'remote_work', 'tech_company']
```

```python
def labelEncoding(df, col):
  le = pp.LabelEncoder()
  df[col] = le.fit_transform(df[col])
  return df
```

```python
def oneHotEncoding(df, col):
  df = pd.get_dummies(df, col)
  return df
```

```python
for i in headers:
  if i not in one_hot_cols and i!='Age':
    df = labelEncoding(df, i)
  else:
    df = labelEncoding(df, i)
# df = oneHotEncoding(df, one_hot_cols)
```

```python
for i in range(len(df['Age'])):
  if(df['Age'][i]>124 or df['Age'][i]<0):
    df['Age'][i]=df['Age'].median()
```

```python
df_x = df.drop(['Timestamp', 'treatment', 'obs_consequence', 'comments'], axis = 1)
df_y = df['treatment']

train_y = np.array(df_y).reshape(1259,1)
train_x = np.array(df_x)

df_y.to_csv('target.csv', index=False)
df_x.to_csv('train.csv', index=False)
```
